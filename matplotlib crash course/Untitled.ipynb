```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,20,20)
x
```

```python
y=np.random.randint(1,15,20)
y
```

```python
plt.plot(y)
```

```python
y=np.sort(y)
```

```python
plt.plot(y)
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('X vs Y')
```

```python
y2=y*x
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r')
plt.subplot(2,1,2)
plt.plot(x,y2,'g')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'ro')
plt.subplot(2,1,2)
plt.plot(x,y2,'g*-')
```

## matlab vs matplotlib

```python
#object oriented method
```

```python
fig=plt.figure()
axes=fig.add_axes([0.5,0.5,1,1])

axes.plot(x,y,'ro-')
axes.set_xlabel('X val')
axes.set_ylabel('Y val')
axes.set_title('random plot')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.5,0.5,1,1])
axes2=fig.add_axes([0.6,1.1,0.3,0.3])

axes1.plot(x,y,'g')
axes2.plot(x,y2,'y')


axes1.set_xlabel('X ')
axes1.set_ylabel('Y')
axes1.set_title('Y plot')


axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Y2 plot')
```

```python
fig,ax=plt.subplots(1,2)
col=['r','g']
data=[y,y2]
for i ,axes in enumerate(ax):
    axes.plot(x,data[i],col[i])
```

```python
fig,ax=plt.subplots(figsize=(8,4))
ax.plot(x,y)
ax.set_xlabel('X ')
ax.set_ylabel('Y')
ax.set_title('Random plot')
```

```python
fig,ax=plt.subplots(figsize=(8,4))
ax.plot(x,y,'r',label='y')
ax.plot(x,y2,'g',label='y2')
ax.set_xlabel('X ')
ax.set_ylabel('Y')
ax.legend()
ax.set_title('Random plot')
```

```python
fig,ax=plt.subplots()
ax.plot(x,y,'ro--',markersize=6)
```

```python
fig,ax=plt.subplots(1,3,figsize=(10,4))
ax[0].plot(x,y,'g',x,y2,'r')

ax[1].plot(x,y2)
ax[2].plot(x,y,x,y**2)
ax[0].set_ylim([0,300])
```
